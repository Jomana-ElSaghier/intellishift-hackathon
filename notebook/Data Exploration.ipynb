```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Display all columns
pd.set_option("display.max_columns", None)
```

```python
# Load dataset
df = pd.read_csv("Data.csv")

print("Dataset Loaded Successfully!")

df.head()
```

```python
print(f"Rows: {df.shape[0]}")
print(f"Columns: {df.shape[1]}")
```

# Dataset Overview

```python
print("\nDataset Shape")
print("-" * 30)
print(f"Rows: {df.shape[0]}")
print(f"Columns: {df.shape[1]}")

print("\nColumn Names")
print("-" * 30)
print(df.columns.tolist())

print("\nData Types")
print("-" * 30)
print(df.dtypes)

print("\nDataset Information")
print("-" * 30)
df.info()

print("\nStatistical Summary")
print("-" * 30)
print(df.describe(include="all"))
```

# Data Quality Assessment

```python
print("\nMissing Values")
print("-" * 30)
print(df.isnull().sum())

print("\nDuplicate Rows")
print("-" * 30)
print(df.duplicated().sum())

print("\nNegative or Zero Sales")
print("-" * 30)
print(len(df[df["Sales"] <= 0]))

print("\nMissing Customer IDs")
print("-" * 30)
print(df["Customer ID"].isnull().sum())

print("\nMissing Product IDs")
print("-" * 30)
print(df["Product ID"].isnull().sum())
```

# Unique Values

```python
print("\nUnique Customers:", df["Customer ID"].nunique())
print("Unique Products:", df["Product ID"].nunique())
print("Unique Orders:", df["Order ID"].nunique())

print("\nRegions")
print(df["Region"].unique())

print("\nCategories")
print(df["Category"].unique())

print("\nShip Modes")
print(df["Ship Mode"].unique())

print("\nCustomer Segments")
print(df["Segment"].unique())
```

# Sales Analysis

```python
print("\nSales Statistics")
print("-" * 30)
print(df["Sales"].describe())

print("\nTop 10 Highest Sales")
print(df.nlargest(10, "Sales"))

print("\nTop 10 Lowest Sales")
print(df.nsmallest(10, "Sales"))

print("\nSales by Category")
print(df.groupby("Category")["Sales"].sum())

print("\nSales by Sub-Category")
print(df.groupby("Sub-Category")["Sales"].sum().sort_values(ascending=False))

print("\nSales by Region")
print(df.groupby("Region")["Sales"].sum())

print("\nTop 10 States by Sales")
print(df.groupby("State")["Sales"].sum().sort_values(ascending=False).head(10))

print("\nTop 10 Customers")
print(df.groupby("Customer Name")["Sales"].sum().sort_values(ascending=False).head(10))

print("\nTop 10 Products")
print(df.groupby("Product Name")["Sales"].sum().sort_values(ascending=False).head(10))
```

# Shipping Analysis

```python
df["Order Date"] = pd.to_datetime(df["Order Date"], dayfirst=True)
df["Ship Date"] = pd.to_datetime(df["Ship Date"], dayfirst=True)

df["Shipping Delay"] = (
    df["Ship Date"] - df["Order Date"]
).dt.days

print("\nShipping Delay Statistics")
print(df["Shipping Delay"].describe())

print("\nAverage Shipping Delay by Ship Mode")
print(df.groupby("Ship Mode")["Shipping Delay"].mean())

print("\nNegative Shipping Delays")
print(len(df[df["Shipping Delay"] < 0]))
```

# Monthly Sales

```python
monthly_sales = (
    df.groupby(df["Order Date"].dt.to_period("M"))["Sales"]
    .sum()
)

print("\nMonthly Sales")
print(monthly_sales)
```

# Business Insights

```python
print("BUSINESS INSIGHTS:")

print("Top Region:")
print(df.groupby("Region")["Sales"].sum().idxmax())

print("\nTop Category:")
print(df.groupby("Category")["Sales"].sum().idxmax())

print("\nTop Customer:")
print(df.groupby("Customer Name")["Sales"].sum().idxmax())

print("\nHighest Revenue State:")
print(df.groupby("State")["Sales"].sum().idxmax())

print("\nAverage Shipping Delay:")
print(round(df["Shipping Delay"].mean(), 2), "days")
```

# Save Processed Dataset
